```python
# Ejercicio N15 Leer un texto y determinar cuántas veces dicho texto contiene cada una de las vocales en minúscula y mayúscula.

texto = input("Ingrese el texto: ")
vocales = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0, 'A': 0, 'E': 0, 'I': 0, 'O': 0, 'U': 0}
for letra in texto:
    if letra in vocales:
        vocales[letra] += 1
for vocal, count in vocales.items():
    print(f"{vocal}: {count}")
```

```python
# Ejercicio N18  Programa que averigüe e imprima cuántos múltiplos de 7, y cuántos múltiplos de 9, hay en los números comprendidos entre 1000 y 5000.

count_seven = 0
count_nine = 0
for num in range(1000, 5001):
    if num % 7 == 0:
        count_seven += 1
    if num % 9 == 0:
        count_nine += 1
print(f"Número de múltiplos de 7 entre 1000 y 5000: {count_seven}")
print(f"Número de múltiplos de 9 entre 1000 y 5000: {count_nine}")
```

```python
# Ejercicio N20 Programa para calcular el factorial de un número.  Consulte previamente qué es el factorial de un número.

num = int(input("Ingresa un número para calcular su factorial: "))
factorial = 1
if num < 0:
    print("No se puede calcular el factorial de un número negativo.")
elif num == 0:
    print("El factorial de 0 es 1")
else:
    for i in range(1, num + 1):
        factorial *= i
    print(f"El factorial de {num} es {factorial}")

```
